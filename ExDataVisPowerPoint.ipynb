```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv("D:\StockageDrive\CV\GoMyCode Cours IA1/tips.csv",sep=",")
```

```python
data
```

```python
data.isnull().sum()
```

```python
grouped_data = data.groupby(['time', 'sex'])['total_bill'].sum().reset_index()

plt.bar(grouped_data['time'] + "-" + grouped_data['sex'], grouped_data['total_bill'],color='orange')
plt.xlabel("Time and Sex")
plt.ylabel("Total Bill Amount")
plt.show()
```

```python
sns.barplot(data=data[['total_bill','tip','size']])
plt.show()
```

```python
plt.scatter(data['total_bill'], data['tip'])
plt.title('Total Bill v/s Total Tip')
plt.xlabel('Total Bill')
plt.ylabel('Total Tip')
plt.show()
```

```python
sns.boxplot(data=data[['total_bill','tip']])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
total_bill_standardized = scaler.fit_transform(data[['total_bill']])
tip_standardized = scaler.fit_transform(data[['tip']])
plt.boxplot([total_bill_standardized.flatten(), tip_standardized.flatten()])
plt.show()
```

```python
sns.displot(data['total_bill'], bins=25, color="orange")
```

```python
sns.catplot(data=data, x="time", y="total_bill", hue="day", kind="bar")
```

```python
sns.catplot(data=data, x="time", y="total_bill", hue="smoker", kind="bar")
```

```python
sns.countplot(x='size',data=data)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.figure(figsize=(5,5))
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(data['total_bill'], color='orange', label='Total Bill')
ax.plot(data['tip'], color='blue', label='Tip')

ax.set_xlabel("Record Number")
ax.set_ylabel("Amount")
ax.legend()
plt.show()
```
